```python
# prompt: buat agar membaca file dari csv dari "C:\Users\rezaprakosta\Downloads\Bike Sales Dataset.csv"

import pandas as pd

# Ganti path dengan path file CSV di Google Colab
file_path = "/content/Bike Sales Dataset.csv"
df = pd.read_csv(file_path)

print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df = df.dropna()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Customer_Age'] = df['Customer_Age'].astype('int')
df['Order_Quantity'] = df['Order_Quantity'].astype('int')
df['Unit_Cost'] = df['Unit_Cost'].astype('float')
df['Unit_Price'] = df['Unit_Price'].astype('float')
df['Profit'] = df['Profit'].astype('float')
df['Cost'] = df['Cost'].astype('float')
df['Revenue'] = df['Revenue'].astype('float')
df['Shipping_Cost'] = df['Shipping_Cost'].astype('float')
```

```python
# Mengonversi kolom tanggal ke format datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.rename(columns={'Date': 'date', 'Day': 'day', 'Month': 'month', 'Year': 'year', 'Customer_ID': 'customer_id',
                   'Customer_Age': 'customer_age', 'Age_Group': 'age_group', 'Customer_Gender': 'customer_gender',
                   'Country': 'country', 'State': 'state', 'Product_Category': 'product_category',
                   'Sub_Category': 'sub_category', 'Product_ID': 'product_id', 'Product': 'product',
                   'Order_ID': 'order_id', 'Order_Quantity': 'order_quantity', 'Unit_Cost': 'unit_cost',
                   'Unit_Price': 'unit_price', 'Profit': 'profit', 'Cost': 'cost', 'Revenue': 'revenue',
                   'Manufacturer': 'manufacturer', 'Shipping_Option': 'shipping_option', 'Shipping_ID': 'shipping_id',
                   'Shipping_Cost': 'shipping_cost', 'Num': 'num'}, inplace=True)
```

```python
import pandas as pd

# Memuat dataset
file_path = "/content/Bike Sales Dataset.csv"
df = pd.read_csv(file_path)

# Dimensi Pelanggan
customer_dim = df[['Age_Group', 'Country', 'Customer_Age', 'Customer_Gender', 'Customer_ID', 'State']].drop_duplicates()

# Dimensi Produk
product_dim = df[['Manufacturer', 'Product', 'Product_Category', 'Product_ID', 'Sub_Category', 'Unit_Cost', 'Unit_Price']].drop_duplicates()

# Dimensi Pengiriman
shipping_dim = df[['Shipping_Cost', 'Shipping_ID', 'Shipping_Option']].drop_duplicates()

# Dimensi Tanggal
date_dim = df[['Date', 'Day', 'Month', 'Year']].drop_duplicates()

# Fakta Pesanan
fact_order = df[['Age_Group', 'Cost', 'Country', 'Customer_Age', 'Customer_Gender', 'Customer_ID', 'Date', 'Day', 'Manufacturer']].drop_duplicates()

# Menyimpan DataFrame ke file CSV
customer_dim.to_csv('customer_dim.csv', index=False)
product_dim.to_csv('product_dim.csv', index=False)
shipping_dim.to_csv('shipping_dim.csv', index=False)
date_dim.to_csv('date_dim.csv', index=False)
fact_order.to_csv('fact_order.csv', index=False)

# Menampilkan beberapa baris pertama dari setiap DataFrame
print("Customer Dimension")
print(customer_dim.head())

print("\nProduct Dimension")
print(product_dim.head())

print("\nShipping Dimension")
print(shipping_dim.head())

print("\nDate Dimension")
print(date_dim.head())

print("\nFact Orders")
print(fact_order.head())
```
